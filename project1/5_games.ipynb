# Анализ рынка видеоигр

**Описание проекта:**   
В интернет-магазине, продающем по всему миру компьютерные игры, имеются  исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимр выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламную кампанию на следующий год.

**Цель проекта:**  
Выявить закономерности, определяющие успешность игры и потенциально популярный продукт, и дать рекомендации для проведения рекламной кампании.

**Задачи проекта:**
- Определить актуальный для проведения анализа период
- Определить наиболее популярные платформы по всему миру и тенденции к изменению
- Определить наиболее прибыльные жанры по всему миру
- Определить для каждого региона (Северная Америка, Европа, Япониия) наиболее прибыльные платформы и жанры, а также оценить - влияние рейтингов на продажи в каждом регионе
- Проверить гипотезу: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"
- Проверить гипотезу: "Средние пользовательские рейтинги жанров Action и Sports  разные"
- Сделать общий вывод

В нашем  распоряжении данные до 2016 года. (данные за 2016 год могут быть неполными!) 

**Описание данных:**

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Столбец 'year_of_release'
Изучим пропуски в столбце:

```python
df[df['year_of_release'].isna()].head()
print('Число пропусков:', df['year_of_release'].isna().sum())
```

Посмотрим есть ли для одинаковых игр год выпуска на других платформах:

```python
df.query('name == "madden nfl 2004"')
df.query('name == "fifa soccer 2004"')
```

Можно увидеть, что в названии некоторых игр стоит год - можно предпололожить, что игра выпускается примерно в том же году, который стоит в названии. Можно использовать это для заполнения пропусков. 

**Это допущение**: игры могут выходить на разных платформахв разное время, но количество пропусков в годах у нас не превышает 1,6% общего кол-ва строк, поэтому вместо удаления таких строк выбираем вариант заполнения.   

**Наблюдение:** в названии некоторых игр содержится год, как правило он на +1 больше реального года выпуска (fifa soccer 2004 - выпущена в 2003), используем это в дальнейшем.

```python
# Заполним пропуски в годах за счет наличия информации об одних и тех же играх, только выпущенных для разных платформ
# если встречаем пропуск, то заполняем его минимальным значением по группе (по названию игры)
for year in df[df['year_of_release'].isna() == True].index:  
    df['year_of_release'][year] = df.loc[df['name'] == df['name'][year], 'year_of_release'].min()
#проверка
df.query('name == "fifa soccer 2004"') 
print('Число пропусков:', df['year_of_release'].isna().sum())
```

Таким образом нам удалось заполнить половину пропусков в столбце год.

В столбце rating содержится рейтинг ESRB. В наборе данных присутствуют следующие значения:
- 'AO' - «Только для взрослых»: Содержание игры только для взрослых старше 18 лет.

- 'EC' - «Для детей младшего возраста»
- 'K-A' - "Для всех" - устаревший, переименован в E
- 'M' - «Для взрослых»: Материалы игры не подходят для подростков младше 17 лет.
- 'RP' - «Рейтинг ожидается»
- 'T' - «Подросткам»: Игра подходит для лиц от 13 лет и старше

ESRB-рейтинг K-A использовался для игр до 1998 года, после был заменен ассоциацией на E - в столбце rating K-A меняем на E


```python
df.loc[df['rating'] == 'K-A','rating'] = 'E'
df['rating'].value_counts()
```

Некоторые значения встречаются совсем редко. Если сравнить их с системой рейтинга, то весьма вероятно, что это неявные дубликаты:
- Рейтинг 'RP' - rating pending - означает, что рейтинга еще нет и игра ожидает его присвоения. Раз рейтинга нет - логично заменить на 'unknown'.
- Рейтинг 'AO' - по сути равен рейтингу M, и то, и то подходит для лиц старше 18.
- Рейтинги 'EC' можно таким же образом включить в рейтинг 'E'.

Все пропущенные значения в столбце rating заменим на unknown: позже это будет необходимо при получении ответа на вопрос о влиянии рейтинга ESRB на продажи в отдельном регионе. 

```python
df.loc[df['rating'] == 'RP', 'rating'] = 'unknown'
df.loc[df['rating'] == 'AO', 'rating'] = 'M'
df.loc[df['rating'] == 'EC', 'rating'] = 'E'
df["rating"].fillna('unknown', inplace = True)
df['rating'].value_counts()
```

Добавим категоризацию по рейтингу для удобства дальнейшего анализа:
    

```python
# функция для категоризации рейтинга
def rating_func(row): 
    rating = row['rating']
    if rating == 'E':
        return 'для всех'
    if rating == 'T':
        return 'от 13 лет и старше'
    if rating == 'M':
        return 'для взрослых'
    if rating == 'E10+':
        return 'от 10 лет и старше'
    if rating == 'unknown':
        return 'нет рейтинга'
```

```python
# Cоздаем новый столбец с применением функции к датафрейму методом apply()
df['rating_cat'] = df.apply(rating_func, axis=1)
df.sample(5) # проверка
```

## Портрет пользователя каждого региона (NA, EU, JP)

   •    Самые популярные платформы (топ-5).   
   •    Самые популярные жанры (топ-5).   
   •    Влияние рейтинга ESRB на продажи в отдельном регионе    

```python
# функция для категоризации по платформам
def top5_platform(row):
    platform = row['platform']
    for column in ['na_sales', 'eu_sales', 'jp_sales']:
        top5region = actual_df.pivot_table(index='platform', values = column, aggfunc = 'sum')\
                              .sort_values(by=column, ascending = False).head()
        if platform in top5region.index:
            return platform
        else:
            return 'Other platform'
```

```python
actual_df['platform_cat'] = actual_df.apply(top5_platform, axis=1)
```

```python
# функция для категоризации по жанрам
def top5_genre(row):
    genre = row['genre']
    for column in ['na_sales', 'eu_sales', 'jp_sales']:
        top5region = actual_df.pivot_table(index='genre', values = column, aggfunc = 'sum')\
                              .sort_values(by=column, ascending = False).head()
        if genre in top5region.index:
            return genre
        else:
            return 'Other genre'
```

```python
actual_df['genre_cat'] = actual_df.apply(top5_genre, axis=1)
```

```python
# функция для прорисовки пирожков
def region_pie(column):
    rows = 1
    cols = 3
    fig, axes = plt.subplots(rows, cols, figsize=(18,5))
    region_list = ['na_sales', 'eu_sales', 'jp_sales']
    title_dict = {'na_sales':'Продажи в Северной Америке',
                  'eu_sales':'Продажи в Европе', 'jp_sales':'Продажи в Японии'}
    count = 0
    for r in range(rows):
        for i in range(cols):
            data = actual_df.pivot_table(index=column, values = region_list[count], aggfunc = 'sum')\
                            .sort_values(by=region_list[count], ascending = False)
            plot = data.plot(kind='pie', subplots=True, ax=axes[i], fontsize=13, legend=False,title=[region_list[count]],\
                            autopct='%1.0f%%', pctdistance=0.8, ylabel=" ")
            plt.ylabel(None)
            count += 1
            
    fig.suptitle(f'Распределение по {column}', y=1); 
```

## Проверка гипотез

### Гипотеза №1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем первую гипотезу:  
**Нулевая гипотеза H0** - средние пользовательские рейтинги платформ Xbox One и PC одинаковые =**не различаются** (средние рейтинги равны).    
**Альтернативная гипотеза H1** - средние пользовательские рейтинги платформ Xbox One и PC не одинаковые = **различаются** (средние рейтинги не равны).   
Если в процессе проведения статистического теста мы отвергаем H0, то тогда принимаем гипотезу H1 (не отвергаем).

```python
# сначала построим диаграмму размаха:
data = actual_df.query('platform == ["XOne", "PC"]')
 
plt.figure(figsize=(8,6))
sns.boxplot(data=data, x='platform', y='user_score', showfliers=False, showmeans=True, 
            meanprops={"marker":"s","markerfacecolor":"white", "markeredgecolor":"blue"})\
            .set(xlabel = 'Платформа', ylabel = 'Оценка пользователей') 
           
plt.title('Средние пользовательские рейтинги платформ Xbox One и PC ');
```

Судя по диаграмме размаха, средние (и медианные) значения отличаются незначительно, поэтому можно предположить, что средние значения пользовательских рейтингов платформ PC и Xbox One практически на одном уровне, проверим это предположение:

Для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, применим метод   
scipy.stats.ttest_ind (array1, array2, equal_var)   
Зададим alpha (критический уровень статистической значимости) - 0.05 (если p-value (уровень значимости) окажется меньше него - отвергнем гипотезу)

```python
#Посчитаем дисперсию обеих выборок:
print('Дисперсия XOne:', actual_df.query('platform == "XOne"')['user_score'].var())
print('Дисперсия PC:', actual_df.query('platform == "PC"')['user_score'].var())
```

```python
#Дисперсии выборок сильно различаются, поэтому задаем значение параметра equal_var = False
alpha = 0.05         
results = st.ttest_ind(actual_df.query('platform == "XOne" & user_score.isna() == False')['user_score'],
                       actual_df.query('platform == "PC" & user_score.isna() == False')['user_score'],
                       equal_var = False)

print('p-значение(уровень значимости):', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу.")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Полученные данные не противоречат H0 гипотезе: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**   
Вероятность получить для данной модели распределения наблюдаемый или еще более далекий от предполагаемого результат = 16% (при критическом уровне статистической значимости 5%).   

Проверим на нашей выборке различаются ли средние значения пользовательских рейтингов платформ Xbox One и PC:

```python
print('Средний user_score платформы Xbox One - {:.2f}'
      .format(actual_df.query('platform == "XOne" & user_score.isna() == False')['user_score'].mean()))
print('Средний user_score платформы PC - {:.2f}'
      .format(actual_df.query('platform == "PC" & user_score.isna() == False')['user_score'].mean()))
```

Разница в 0.24 балла при оценке платформ пользователями не является статистически значимой.
